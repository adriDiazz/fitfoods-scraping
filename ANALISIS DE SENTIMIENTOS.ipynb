```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
from pandas import json_normalize
```

```python
df = pd.read_csv('./commentsFinal.csv', sep=';')
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
sentimientos = []
for texto in df['text']:
    if isinstance(texto, str):
        sentiment = analyzer.polarity_scores(texto)
        analisis = sentiment['compound']
        sentimientos.append(analisis)
    
df['sentiment'] = sentimientos
```

```python
df.to_csv('commentsAnalized.csv')
```
