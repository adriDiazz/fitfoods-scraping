```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
from pandas import json_normalize
```

```python
df = pd.read_csv('./commentsFinal.csv', sep=';')
```

```python
df
```

```python
from googletrans import Translator
import time

translator = Translator(service_urls=[
      'translate.google.com'])
```

```python
for index, row in df.iterrows():
    comment = row['text']
    if comment is not None and comment != '':
        try:
            translated_comment = translator.translate(comment, dest='en')
            df.at[index, 'comentario_en'] = translated_comment.text
        except Exception as e:
            print(f"Error translating row {index}: {e}")
    time.sleep(0.1)
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
sentimientos = []
for texto in df['comentario_en']:
    if isinstance(texto, str):
        sentiment = analyzer.polarity_scores(texto)
        analisis = sentiment['compound']
        sentimientos.append(analisis)
    else:
        sentimientos.append(0)
    
df['sentiment'] = sentimientos
```

```python
df.drop('comentario_en', axis=1, inplace=True)
```
