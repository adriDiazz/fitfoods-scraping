```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import json
from pandas import json_normalize
```

```python
ruta_archivo_json = './comentarios.json'
```

```python
df = pd.read_json(ruta_archivo_json, orient ='index')
```

```python
analyzer = SentimentIntensityAnalyzer()
sentimientos = []
for texto in df['comentario']:
    sentiment = analyzer.polarity_scores(texto)
    analisis = sentiment['compound']
    sentimientos.append(analisis)
    
df['sentimientos'] = sentimientos
```
